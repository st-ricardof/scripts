```python
#Atualizo o  DataFrame apenas com as colunas que possuem valores
df_cat_foto = pd.DataFrame(
    {'nreferencia':df_cat_foto['nreferencia'], 
     'especificação':df_cat_foto['especificação'], 
     'numero':df_cat_foto['numero'],
     'quantidade':df_cat_foto['quantidade'],
     'sistema':df_cat_foto['Sistema'],
     'titulo':df_cat_foto['titulo'],
     'local':df_cat_foto['local'],
     'data':df_cat_foto['data'],
     'produção':df_cat_foto['produção']
    })
```

```python
df_cat_foto
```

```python
#verifico os tipos de valores do DF
df_cat_foto.dtypes
```

```python
df_cat_foto.groupby('quantidade')['quantidade'].agg([np.size])
#OK, quantidade não é inteiro mesmo
```

```python
df_cat_foto.local = df_cat_foto.local.replace('CAPH', 'CAPH-USP')
df_cat_foto.local = df_cat_foto.local.replace('São Paulo/Anfiteatro de Geografia', 'São Paulo/Anfiteatro Geografia')
df_cat_foto.groupby('local')['local'].agg([np.size])
```

```python
df_cat_foto.data = df_cat_foto.data.replace('s/d', None)
df_cat_foto.groupby('data')['data'].agg([np.size])
```

```python
df_cat_foto.groupby('produção')['produção'].agg([np.size])
```
